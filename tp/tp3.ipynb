```python
""" ecrire une fonction qui prend en entré r la distance entre les deux corps m1, m2 et qui construit la courbe F en fonction de r"""

m1 = 1.5
m2 = 0.5
G = 6.674 *(10**-11)
listf = []
listr = range(100,1001,50)
from matplotlib.pyplot import *
    
    
```

```python
def F(r):
    f = (G*m1*m2)/(r**2)
    return f
```

```python
""" fonction range pas 0.0.01 """
def rangeTime(init,f,p):
    time = [init]
    i = p
    while time[-1] < f:
        time.append(i)
        i = i + p
    return time
rangeTime(0,1,0.1)
```

```python
from math import sin
from math import cos
from matplotlib.pyplot import *
def draw_graph(x,y):
    plot(x,y,marker='o')
    xlabel('coordinate x')
    ylabel('coordinate y')
    title('Projectile motion of ball')
def generate(u,teta):
    g = 9.8
    t_last  = 2*((u*sin(teta))/g)
    rt = rangeTime(0,t_last,0.001)
    
    
    sx = []
    sy = []
    for j in rt:
        s1 = u*cos(teta)*j
        sx.append(s1)
        s2 = (u*sin(teta)*j) - (0.5*g*(j**2))
        sy.append(s2)
    draw_graph(sx,sy)
generate(5,45)
        
```

```python
import pylab as pl
import numpy as np 
""" create a figure of size 9x7 inches, 100 dots per inch"""
pl.figure(figsize=(9,7),dpi=100)
""" create a new subplot from a grid 1x1""" 
pl.subplot(1,1,1)
"""  difine x aixs begin to end with 10000 pt"""
X = np.linspace(-np.pi*2,np.pi*2,10e4)

S1,S2 = np.sin(X),np.sin(2*X)

pl.plot(X,S1, color='blue', linewidth=1.0, linestyle="-",label="$sin(x)$")
pl.plot(X,S2, color='red', linewidth=1.0, linestyle="-",label="$sin(2x)$")
""" set x limits from -2.5*pi to 2.5*pi"""
pl.xlim(-2.5*np.pi,2.5*np.pi)

""" set x ticks """
pl.xticks(np.linspace(-2.5*np.pi,2.5*np.pi,9))
""" set y limits"""
pl.ylim(-1,2,1.2)
""" set y ticks"""
pl.yticks(np.linspace(-1,1,5))
""" set the tile as sine waves """
pl.title('$sin(x) and sin(2x) waves')
""" setting labels"""
pl.xlabel('$x$')
pl.ylabel('$sin(x) and sin(2x)$')
pl.grid(True)
pl.legend()
```
